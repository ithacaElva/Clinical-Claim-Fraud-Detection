```python
def sequence_len(claim_data): #1~500 claim_data in claimdata
    key_list = claim_data['resource'].keys()
    if 'item' in key_list: 
        return len(claim_data['resource']['item'])
    else: 
        return 0
```

```python
def bill_start_end(claim_data): 
    key_list = claim_data['resource'].keys()
    if 'billablePeriod' in key_list:
        start = claim_data['resource']['billablePeriod']['start'] #'2014-01-13T19:17:38+01:00'
        end = claim_data['resource']['billablePeriod']['end']
        if len(start) == 25:
          start_timestamp = time.mktime(time.strptime(start[0:10] + ' ' + start[11:19], '%Y-%m-%d %H:%M:%S'))
          end_timestamp = time.mktime(time.strptime(end[0:10] + ' ' + end[11:19], '%Y-%m-%d %H:%M:%S'))
          return start_timestamp, end_timestamp
        else:
          return 0,0
    else:
        return 0,0
```

```python
def provider(claim_data):
    key_list = claim_data['resource'].keys()
    if 'provider' in key_list:
        key_list2 = claim_data['resource']['provider'].keys()
        if 'reference' and 'display' in key_list2:
            return claim_data['resource']['provider']['reference'], claim_data['resource']['provider']['display']
        elif 'reference' in key_list2 and 'display' not in key_list2:
            return claim_data['resource']['provider']['reference'], 'None'
        elif 'reference' not in key_list2 and 'display' in key_list2:
            return 'None', claim_data['resource']['provider']['display']
        else:
            return 'None', 'None'
    else:
        return 'None', 'None'
```

```python
def facility(claim_data):
    key_list = claim_data['resource'].keys()
    if 'facility' in key_list:
        key_list2 = claim_data['resource']['facility'].keys()
        if 'reference' and 'display' in key_list2:
            return claim_data['resource']['facility']['reference'], claim_data['resource']['facility']['display']
        elif 'reference' in key_list2 and 'display' not in key_list2:
            return claim_data['resource']['facility']['reference'], 'None'
        elif 'reference' not in key_list2 and 'display' in key_list2:
            return 'None', claim_data['resource']['facility']['display']
        else:
            return 'None', 'None'
    else:
        return 'None', 'None' 
```

```python
def insurance_company(claim_data):
    key_list = claim_data['resource'].keys()
    if 'insurance' in key_list:
        key_list2 = claim_data['resource']['insurance'][0].keys()
        if 'coverage' in key_list2:
            key_list3 = claim_data['resource']['insurance'][0]['coverage'].keys()
            if 'display' in key_list3:
                return claim_data['resource']['insurance'][0]['coverage']['display']
            else:
                return 'No Insurance'
        else:
                return 'No Insurance'
    else:
        return 'No Insurance'
```

```python
def total_payment(claim_data):
    key_list = claim_data['resource'].keys()
    if 'total' in key_list:
        return claim_data['resource']['total']['value']
    else:
        return 'None'
```

```python
def providertype(claim_data):
  key_list = claim_data['resource'].keys()
  if 'type' in key_list:
    return claim_data['resource']['type']['coding'][0]['code']
  else:
    return 'None'
```

```python
def patient(claim_data):
  key_list = claim_data['resource'].keys()
  if 'patient' in key_list:
    key_list2 = claim_data['resource']['patient'].keys()
    if 'reference' in key_list2:
      return claim_data['resource']['patient']['reference'][8::]
    else:
      return 'None'
  else:
    return 'None'
```

```python
def priority_r(claim_data):
  key_list = claim_data['resource'].keys()
  if 'priority' in key_list:
    return claim_data['resource']['priority']['coding'][0]['code']
  else:
    return 'None'
```

```python
def encounter(claim_data):
  key_list = claim_data['resource'].keys()
  if 'item' in key_list:
    key_list2 = claim_data['resource']['item'][0]
    if 'productOrService' in key_list2:
      key_list3 = claim_data['resource']['item'][0]['productOrService'].keys()
      if 'coding' in key_list3:
        key_list4 = claim_data['resource']['item'][0]['productOrService']['coding'][0].keys()
        if 'display' in key_list4:
          return claim_data['resource']['item'][0]['productOrService']['coding'][0]['display']
        else:
          return 'None'
      else:
        return 'None'
    else:
      return 'None'
  else:
    return 'None'

#def sequence_len(claim_data): #1~500 claim_data in claimdata
#    key_list = claim_data['resource'].keys()
#    if 'item' in key_list: 
#        return len(claim_data['resource']['item'])
#    else: 
#        return 0
```

```python
def claim_clean(claim_data): #1~500 claim_data in claimdata
    claim_type = claim_data['resource']['resourceType']
    claim_id = claim_data['resource']['id']
    last_update_date = claim_data['resource']['meta']['lastUpdated']
    provider_type = providertype(claim_data)#claim_data['resource']['type']['coding'][0]['code']
    patient_id = patient(claim_data)#claim_data['resource']['patient']['reference']
    billable_start = bill_start_end(claim_data)[0] #claim_data['resource']['billablePeriod']['start']
    billable_end = bill_start_end(claim_data)[1]   #claim_data['resource']['billablePeriod']['end']
    billable_time_difference = billable_end - billable_start    #timestamp: 时间差
    provider_id = provider(claim_data)[0]  #claim_data['resource']['provider']['reference']
    provider_name = provider(claim_data)[1]  #claim_data['resource']['provider']['display']
    priority = priority_r(claim_data) #claim_data['resource']['priority']['coding'][0]['code']
    facility_location_id= facility(claim_data)[0] #claim_data['facility']['reference']
    facility_location_description = facility(claim_data)[1] #claim_data['facility']['display']
    insurancecompany = insurance_company(claim_data)#claim_data['resource']['insurance'][0]['coverage']['display']
    num_sequence = sequence_len(claim_data)
    totalpayment = total_payment(claim_data)#claim_data['resource']['total']
    encounter_des = encounter(claim_data)

    
    return claim_type, claim_id, last_update_date, provider_type, patient_id, billable_start, billable_end, billable_time_difference, provider_id, provider_name, priority, facility_location_id, facility_location_description, insurancecompany, num_sequence, totalpayment, encounter_des
#'sequence': 'net', 'text',
```
